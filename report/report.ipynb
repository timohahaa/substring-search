# Эксперимент 1: Влияние длины подстроки на время работы
В этом эксперименте длина строки не меняется, а длина подстроки изменяется от 1 до 1 000 (в случае с текстом Lorem до 445). Теоретически, поскольку каждый алгоритм работает за линейное время (кроме наивной реализации) влияние n будет тем больше, чем ближе длина подстроки к длине самой строки. 
Запуск на Aurus 15 XE (Core i7-12700H) дал следующие результаты:

```python
lorem = pd.read_csv("../resources/lorem_pattern_length.csv")
dna = pd.read_csv("../resources/dna_pattern_length.csv")
alice = pd.read_csv("../resources/alice_pattern_length.csv")
#aurelius = pd.read_csv("../resources/aurelius_pattern_length.csv")

#lorem.sort_values('length', inplace=True)
#dna.sort_values('length', inplace=True)
#alice.sort_values('length', inplace=True)
#aurelius.sort_values('length', inplace=True)
```

```python
lorem.plot(
    x='length', 
    y=['Go', 'Naive', 'RK', 'BM', 'KMP'],
    title='lorem', 
    xlabel='pattern length', 
    ylabel='time (nanoseconds)',
)
```

```python
dna.plot(
    x='length', 
    y=['Go', 'Naive', 'RK', 'BM', 'KMP'],
    title='dna', 
    xlabel='pattern length', 
    ylabel='time (nanoseconds)',
)
```

```python
alice.plot(
    x='length', 
    y=['Go', 'Naive', 'RK', 'BM', 'KMP'],
    title='alice', 
    xlabel='pattern length', 
    ylabel='time (nanoseconds)',
)
```
